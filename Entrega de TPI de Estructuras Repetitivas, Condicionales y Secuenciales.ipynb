#### **Ejercicio 1 - Caja del Kiosco**

```python
# 1. Pedir nombre del cliente
nombre = input("Cliente: ").strip()

while nombre == "" or not nombre.isalpha():
    print("Error. Ingrese un nombre válido, solo con letras.")
    nombre = input("Cliente: ").strip()

# 2. Pedir cantidad de productos
cantidad = input("Cantidad de productos: ").strip()

while not cantidad.isdigit() or int(cantidad) <= 0:
    print("Error. Ingrese un número entero positivo mayor que 0.")
    cantidad = input("Cantidad de productos: ").strip()

cantidad = int(cantidad)

# Variables acumuladoras
total_sin_descuentos = 0
total_con_descuentos = 0

# 3. Recorrer cada producto
for i in range(1, cantidad + 1):
    precio = input(f"Producto {i} - Precio: ").strip()

    while not precio.isdigit():
        print("Error. Ingrese un precio válido en números enteros.")
        precio = input(f"Producto {i} - Precio: ").strip()

    precio = int(precio)
    total_sin_descuentos += precio

    descuento = input(f"Producto {i} - Descuento (S/N): ").strip().lower()

    while descuento != "s" and descuento != "n":
        print("Error. Debe ingresar S o N.")
        descuento = input(f"Producto {i} - Descuento (S/N): ").strip().lower()

    if descuento == "s":
        precio_final = precio * 0.90
    else:
        precio_final = precio

    total_con_descuentos += precio_final

# 4. Cálculos finales
ahorro_total = total_sin_descuentos - total_con_descuentos
promedio = total_con_descuentos / cantidad

# Mostrar resultados
print(f"\nTotal sin descuentos: ${total_sin_descuentos}")
print(f"Total con descuentos: ${total_con_descuentos:.2f}")
print(f"Ahorro: ${ahorro_total:.2f}")
print(f"Promedio por producto: ${promedio:.2f}")
```

#### **Ejercicio 2 - Acceso al Campus y Menú Seguro**

```python
# 1. Credenciales fijas
usuario_correcto = "alumno"
clave_correcta = "python123"

intentos = 0
acceso = False

# 2. Sistema de login (máximo 3 intentos)
while intentos < 3 and not acceso:
    print(f"\nIntento {intentos + 1}/3")
    
    usuario = input("Usuario: ").strip()
    clave = input("Clave: ").strip()
    
    if usuario == usuario_correcto and clave == clave_correcta:
        print("Acceso concedido.")
        acceso = True
    else:
        print("Error: credenciales inválidas.")
        intentos += 1

# 3. Bloqueo si falla
if not acceso:
    print("Cuenta bloqueada.")
else:
    # 4. Menú repetitivo
    salir = False
    
    while not salir:
        print("\n1) Estado  2) Cambiar clave  3) Mensaje  4) Salir")
        opcion = input("Opción: ").strip()
        
        # 5. Validación del menú
        if not opcion.isdigit():
            print("Error: ingrese un número válido.")
            continue
        
        opcion = int(opcion)
        
        if opcion < 1 or opcion > 4:
            print("Error: opción fuera de rango.")
            continue
        
        # Opciones
        if opcion == 1:
            print("Estado: Inscripto")
        
        elif opcion == 2:
            # Cambio de clave
            nueva_clave = input("Nueva clave: ").strip()
            
            if len(nueva_clave) < 6:
                print("Error: mínimo 6 caracteres.")
                continue
            
            confirmación = input("Confirmar clave: ").strip()
            
            if nueva_clave != confirmación:
                print("Error: las claves no coinciden.")
            else:
                clave_correcta = nueva_clave
                print("Clave actualizada correctamente.")
        
        elif opcion == 3:
            print("Sigue adelante, cada línea de código te acerca a tu meta.")
        
        elif opcion == 4:
            print("Sesión finalizada.")
            salir = True
```

#### **Ejercicio 3 - Agenda de Turnos con Nombres**

```python
# 1. Nombre del operador
operador = input("Nombre del operador: ").strip()

while operador == "" or not operador.isalpha():
    print("Error. Ingrese solo letras.")
    operador = input("Nombre del operador: ").strip()

# Turnos (vacíos al inicio)
lunes1 = ""
lunes2 = ""
lunes3 = ""
lunes4 = ""

martes1 = ""
martes2 = ""
martes3 = ""

salir = False

# 2. Menú principal
while not salir:
    print("\n1) Reservar  2) Cancelar  3) Ver agenda  4) Resumen  5) Salir")
    opcion = input("Opción: ").strip()

    if not opcion.isdigit():
        print("Error: ingrese un número válido.")
        continue

    opcion = int(opcion)

    if opcion < 1 or opcion > 5:
        print("Error: opción fuera de rango.")
        continue

    # -----------------------------------
    # 1. RESERVAR TURNO
    # -----------------------------------
    if opcion == 1:
        dia = input("Día (1=Lunes, 2=Martes): ").strip()

        while not dia.isdigit() or int(dia) not in (1, 2):
            print("Error. Ingrese 1 o 2.")
            dia = input("Día (1=Lunes, 2=Martes): ").strip()

        nombre = input("Nombre del paciente: ").strip()

        while nombre == "" or not nombre.isalpha():
            print("Error. Solo letras.")
            nombre = input("Nombre del paciente: ").strip()

        reservado = False

        # LUNES
        if dia == "1":
            # Verificar repetido
            if nombre == lunes1 or nombre == lunes2 or nombre == lunes3 or nombre == lunes4:
                print("Error: paciente ya tiene turno ese día.")
            else:
                if lunes1 == "":
                    lunes1 = nombre
                    reservado = True
                elif lunes2 == "":
                    lunes2 = nombre
                    reservado = True
                elif lunes3 == "":
                    lunes3 = nombre
                    reservado = True
                elif lunes4 == "":
                    lunes4 = nombre
                    reservado = True
                else:
                    print("No hay turnos disponibles el Lunes.")

        # MARTES
        else:
            if nombre == martes1 or nombre == martes2 or nombre == martes3:
                print("Error: paciente ya tiene turno ese día.")
            else:
                if martes1 == "":
                    martes1 = nombre
                    reservado = True
                elif martes2 == "":
                    martes2 = nombre
                    reservado = True
                elif martes3 == "":
                    martes3 = nombre
                    reservado = True
                else:
                    print("No hay turnos disponibles el Martes.")

        if reservado:
            print("Turno reservado correctamente.")

    # -----------------------------------
    # 2. CANCELAR TURNO
    # -----------------------------------
    elif opcion == 2:
        dia = input("Día (1=Lunes, 2=Martes): ").strip()

        while not dia.isdigit() or int(dia) not in (1, 2):
            print("Error. Ingrese 1 o 2.")
            dia = input("Día (1=Lunes, 2=Martes): ").strip()

        nombre = input("Nombre del paciente: ").strip()

        while nombre == "" or not nombre.isalpha():
            print("Error. Solo letras.")
            nombre = input("Nombre del paciente: ").strip()

        encontrado = False

        if dia == "1":
            if lunes1 == nombre:
                lunes1 = ""
                encontrado = True
            elif lunes2 == nombre:
                lunes2 = ""
                encontrado = True
            elif lunes3 == nombre:
                lunes3 = ""
                encontrado = True
            elif lunes4 == nombre:
                lunes4 = ""
                encontrado = True

        else:
            if martes1 == nombre:
                martes1 = ""
                encontrado = True
            elif martes2 == nombre:
                martes2 = ""
                encontrado = True
            elif martes3 == nombre:
                martes3 = ""
                encontrado = True

        if encontrado:
            print("Turno cancelado.")
        else:
            print("Paciente no encontrado.")

    # -----------------------------------
    # 3. VER AGENDA DEL DÍA
    # -----------------------------------
    elif opcion == 3:
        dia = input("Día (1=Lunes, 2=Martes): ").strip()

        while not dia.isdigit() or int(dia) not in (1, 2):
            print("Error. Ingrese 1 o 2.")
            dia = input("Día (1=Lunes, 2=Martes): ").strip()

        if dia == "1":
            print("\n--- Lunes ---")
            print("Turno 1:", lunes1 if lunes1 != "" else "(libre)")
            print("Turno 2:", lunes2 if lunes2 != "" else "(libre)")
            print("Turno 3:", lunes3 if lunes3 != "" else "(libre)")
            print("Turno 4:", lunes4 if lunes4 != "" else "(libre)")
        else:
            print("\n--- Martes ---")
            print("Turno 1:", martes1 if martes1 != "" else "(libre)")
            print("Turno 2:", martes2 if martes2 != "" else "(libre)")
            print("Turno 3:", martes3 if martes3 != "" else "(libre)")

    # -----------------------------------
    # 4. RESUMEN GENERAL
    # -----------------------------------
    elif opcion == 4:
        ocupados_lunes = 0
        ocupados_martes = 0

        if lunes1 != "": ocupados_lunes += 1
        if lunes2 != "": ocupados_lunes += 1
        if lunes3 != "": ocupados_lunes += 1
        if lunes4 != "": ocupados_lunes += 1

        if martes1 != "": ocupados_martes += 1
        if martes2 != "": ocupados_martes += 1
        if martes3 != "": ocupados_martes += 1

        disponibles_lunes = 4 - ocupados_lunes
        disponibles_martes = 3 - ocupados_martes

        print("\n--- RESUMEN ---")
        print(f"Lunes -> Ocupados: {ocupados_lunes} | Disponibles: {disponibles_lunes}")
        print(f"Martes -> Ocupados: {ocupados_martes} | Disponibles: {disponibles_martes}")

        if ocupados_lunes > ocupados_martes:
            print("Día con más turnos: Lunes")
        elif ocupados_martes > ocupados_lunes:
            print("Día con más turnos: Martes")
        else:
            print("Empate entre ambos días")

    # -----------------------------------
    # 5. SALIR
    # -----------------------------------
    elif opcion == 5:
        print("Sistema cerrado.")
        salir = True
```

#### **Ejercicio 4 - Escape Room: La Bóveda**

```python
# Variables iniciales
energia = 100
tiempo = 12
cerraduras_abiertas = 0
alarma = False
codigo_parcial = ""

# Control de spam (forzar seguidas)
forzar_seguidas = 0

# Nombre del agente
agente = input("Nombre del agente: ").strip()

while agente == "" or not agente.isalpha():
    print("Error. Ingrese solo letras.")
    agente = input("Nombre del agente: ").strip()

# Bucle principal
while energia > 0 and tiempo > 0 and cerraduras_abiertas < 3:
    
    # Regla de bloqueo por alarma
    if alarma and tiempo <= 3:
        print("\nEl sistema se bloquea por la alarma activa...")
        break

    # Estado actual
    print("\n===== ESTADO =====")
    print(f"Energía: {energia}")
    print(f"Tiempo: {tiempo}")
    print(f"Cerraduras abiertas: {cerraduras_abiertas}/3")
    print(f"Alarma: {'ACTIVA' if alarma else 'OFF'}")
    print(f"Código parcial: {codigo_parcial}")

    # Menú
    print("\n1) Forzar cerradura  2) Hackear panel  3) Descansar")
    opcion = input("Opción: ").strip()

    # Validación
    if not opcion.isdigit():
        print("Error: ingrese un número válido.")
        continue

    opcion = int(opcion)

    if opcion < 1 or opcion > 3:
        print("Error: opción fuera de rango.")
        continue

    # -----------------------------------
    # 1. FORZAR CERRADURA
    # -----------------------------------
    if opcion == 1:
        energia -= 20
        tiempo -= 2
        forzar_seguidas += 1

        # Regla anti-spam
        if forzar_seguidas == 3:
            print("La cerradura se trabó por forzar demasiado...")
            alarma = True
            continue

        # Riesgo si energía baja
        if energia < 40:
            print("Riesgo de activar alarma.")
            riesgo = input("Elegí un número (1-3): ").strip()

            while not riesgo.isdigit() or int(riesgo) < 1 or int(riesgo) > 3:
                print("Error. Ingrese 1, 2 o 3.")
                riesgo = input("Elegí un número (1-3): ").strip()

            if riesgo == "3":
                print("¡Alarma activada!")
                alarma = True
                continue

        # Si no hay alarma, abre cerradura
        if not alarma:
            cerraduras_abiertas += 1
            print("Cerradura abierta con éxito.")

    # -----------------------------------
    # 2. HACKEAR PANEL
    # -----------------------------------
    elif opcion == 2:
        energia -= 10
        tiempo -= 3
        forzar_seguidas = 0  # corta racha

        print("Hackeando sistema...")

        for i in range(4):
            codigo_parcial += "A"
            print(f"Progreso: {codigo_parcial}")

        # Si alcanza longitud suficiente
        if len(codigo_parcial) >= 8 and cerraduras_abiertas < 3:
            cerraduras_abiertas += 1
            print("¡Código completo! Se abre una cerradura.")

    # -----------------------------------
    # 3. DESCANSAR
    # -----------------------------------
    elif opcion == 3:
        tiempo -= 1
        forzar_seguidas = 0  # corta racha

        energia += 15
        if energia > 100:
            energia = 100

        if alarma:
            energia -= 10
            print("La alarma consume energía extra...")

        print("Descansaste un momento.")

# -----------------------------------
# RESULTADO FINAL
# -----------------------------------
print("\n===== RESULTADO =====")

if cerraduras_abiertas == 3:
    print("¡Victoria! Abriste la bóveda.")
elif energia <= 0 or tiempo <= 0:
    print("Derrota. Te quedaste sin recursos.")
elif alarma and tiempo <= 3:
    print("Derrota. El sistema se bloqueó por la alarma.")
else:
    print("Fin del juego.")
```

#### **Ejercicio 5 - Escape Room: La Arena del Gladiador**

```python
print("\n--- BIENVENIDO A LA ARENA ---")

# Paso 1: Nombre del gladiador
nombre = input("Nombre del Gladiador: ").strip()

while nombre == "" or not nombre.isalpha():
    print("Error: Solo se permiten letras.")
    nombre = input("Nombre del Gladiador: ").strip()

# Paso 2: Inicialización
vida_jugador = 100
vida_enemigo = 100
pociones = 3

danio_pesado = 15
danio_enemigo = 12

turno_jugador = True
juego_activo = True

print("\n=== INICIO DEL COMBATE ===")

# Paso 3: Ciclo de combate
while vida_jugador > 0 and vida_enemigo > 0:

    print(f"\n{nombre} (HP: {vida_jugador}) vs Enemigo (HP: {vida_enemigo}) | Pociones: {pociones}")
    print("Elige acción:")
    print("1. Ataque Pesado")
    print("2. Ráfaga Veloz")
    print("3. Curar")

    opcion = input("Opción: ").strip()

    # Validación
    if not opcion.isdigit():
        print("Error: Ingrese un número válido.")
        continue

    opcion = int(opcion)

    if opcion < 1 or opcion > 3:
        print("Error: opción fuera de rango.")
        continue

    # -----------------------------------
    # ATAQUE PESADO
    # -----------------------------------
    if opcion == 1:
        if vida_enemigo < 20:
            danio = danio_pesado * 1.5  # float
            print("¡Golpe Crítico!")
        else:
            danio = danio_pesado

        vida_enemigo -= danio
        print(f"¡Atacaste al enemigo por {danio:.2f} puntos de daño!")

    # -----------------------------------
    # RÁFAGA VELOZ
    # -----------------------------------
    elif opcion == 2:
        print(">> ¡Inicias una ráfaga de golpes!")

        for i in range(3):
            vida_enemigo -= 5
            print("> Golpe conectado por 5 de daño")

    # -----------------------------------
    # CURAR
    # -----------------------------------
    elif opcion == 3:
        if pociones > 0:
            vida_jugador += 30
            pociones -= 1

            if vida_jugador > 100:
                vida_jugador = 100

            print("Te curaste 30 puntos de vida.")
        else:
            print("¡No quedan pociones!")

    # -----------------------------------
    # TURNO DEL ENEMIGO
    # -----------------------------------
    if vida_enemigo > 0:
        vida_jugador -= danio_enemigo
        print(f">> ¡El enemigo contraataca por {danio_enemigo} puntos!")

print("\n=== FIN DEL COMBATE ===")

# Paso 4: Resultado final
if vida_jugador > 0:
    print(f"¡VICTORIA! {nombre} ha ganado la batalla.")
else:
    print("DERROTA. Has caído en combate.")
```
